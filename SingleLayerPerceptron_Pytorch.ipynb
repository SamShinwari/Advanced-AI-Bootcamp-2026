#Single Layer Perceptron in Pytorch

###In this code i used the Pytorch library for Single Layer Perceptron
####I used the AND gate input data

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Input features
X = torch.tensor([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
])

# Target labels
y = torch.tensor([
    [0.],
    [0.],
    [0.],
    [1.]
])
```

```python
class Perceptron(nn.Module):
    def __init__(self):
        super(Perceptron, self).__init__()

        self.linear = nn.Linear(2, 1)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        x = self.linear(x)
        x = self.activation(x)
        return x
```

```python
model = Perceptron()
```

```python
criterion = nn.BCELoss()           # Binary Cross Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
epochs = 1000

for epoch in range(epochs):

    # Forward pass
    outputs = model(X)

    # Compute loss
    loss = criterion(outputs, y)

    # Backpropagation
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():

    outputs = model(X)

    predictions = (outputs >= 0.5).float()

    print("\nPredictions:")
    print(predictions)

    print("\nActual:")
    print(y)
```

## i view learned parameter

```python
print("\nWeights:")
print(model.linear.weight)

print("\nBias:")
print(model.linear.bias)
```
