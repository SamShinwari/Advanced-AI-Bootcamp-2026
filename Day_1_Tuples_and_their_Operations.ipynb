# Day 1: Python Sets for AI, NLP, LLMs & Agentic AI

## What is a Set?
A **set** is an **unordered**, **mutable**, and **unique** collection of items in Python. Unlike lists and tuples, a set **does not allow duplicate values**. Sets are commonly used in AI and NLP for removing duplicates, finding common elements, vocabulary creation, and fast membership testing.

**Syntax:**

```python
my_set = {item1, item2, item3}
```

## Key Features
* Unordered collection
* Mutable (can add or remove items)
* Does not allow duplicate values
* Fast membership testing (`in`)
* Supports mathematical operations like union and intersection

# Real AI Applications

## Example 1: Remove Duplicate Words

````python
sentence = "AI AI is transforming the world"

words = sentence.split()

unique_words = set(words)

print(unique_words)
````

### Application
Text preprocessing removes repeated words before analysis.

---

## Example 2: Build Vocabulary

````python
documents = [
    "I love AI",
    "AI is amazing",
    "Python for AI"
]

vocabulary = set()

for sentence in documents:
    vocabulary.update(sentence.split())

print(vocabulary)
````

### Application
Create a vocabulary of unique words for NLP models.

---

## Example 3: Find Common Words

````python
sentence1 = {"AI", "Python", "Machine Learning"}

sentence2 = {"Python", "Deep Learning", "AI"}

common = sentence1.intersection(sentence2)

print(common)
````

### Application
Find shared keywords between documents.

---

## Example 4: Combine Two Vocabularies

````python
vocab1 = {"AI", "Python"}

vocab2 = {"NLP", "Python", "LLM"}

combined = vocab1.union(vocab2)

print(combined)
````

### Application
Merge vocabularies collected from multiple datasets.

---

## Example 5: Remove Stop Words

````python
words = {
    "I",
    "love",
    "AI",
    "is",
    "powerful"
}

stop_words = {
    "I",
    "is"
}

filtered = words - stop_words

print(filtered)
````

### Application
Remove stop words during NLP preprocessing.

---

## Example 6: Detect Duplicate Documents

````python
documents = set()

documents.add("Introduction to AI")
documents.add("Machine Learning")
documents.add("Introduction to AI")

print(documents)
````

### Application
Prevent duplicate documents from being indexed.

````python
team1 = {"Python", "PyTorch", "Docker"}

team2 = {"Python", "Git", "Docker"}

print(team1.intersection(team2))
````

Find common technologies used by two AI teams.

## Difference

````python
available = {"Python", "PyTorch", "TensorFlow"}

required = {"Python", "TensorFlow"}

print(available - required)
````

# Why Use Sets Instead of Lists?

| List                      | Set                          |
| :------------------------ | :--------------------------- |
| Allows duplicates         | No duplicates                |
| Ordered                   | Unordered                    |
| Slower membership testing | Very fast membership testing |
| Good for sequences        | Good for unique collections  |
| Example: Training samples | Example: Vocabulary          |
