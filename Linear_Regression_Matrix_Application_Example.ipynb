### 1. Problem Description: House Price Prediction with Linear Regression

We want to predict house prices based on their size using a simple linear model. The given data points are:

| House | Size (m²) | Price ($1000s) |
| :---- | :-------- | :------------- |
| 1     | 50        | 150            |
| 2     | 70        | 210            |

The linear model is expressed as: `Price = w1 * Size + b`, where `w1` is the coefficient and `b` is the intercept. We need to find the values of `w1` and `b` that best fit our data.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data
data = {
    'Size (m²)': [50, 70],
    'Price ($1000s)': [150, 210]
}
df = pd.DataFrame(data)

print("Original Data:")
display(df)
```

### 2. Implementing Linear Regression with Scikit-learn

Scikit-learn is a powerful library for machine learning in Python. We will use its `LinearRegression` model to find the optimal `w1` and `b` values. The model will be trained on our `Size` data to predict `Price`.

```python
# Reshape the 'Size' feature to be 2D as required by scikit-learn
X = df[['Size (m²)']].values
y = df['Price ($1000s)'].values

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Extract the coefficient (w1) and intercept (b)
w1 = model.coef_[0]
b = model.intercept_

print(f"Coefficient (w1): {w1:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"The fitted linear model is: Price = {w1:.2f} * Size + {b:.2f}")
```

### 3. Making Predictions

Once the model is trained, we can use it to predict prices for new house sizes or to see how well it fits our existing data. Let's predict the prices for our original data points.

```python
# Make predictions on the training data
predictions = model.predict(X)
df['Predicted Price ($1000s)'] = predictions

print("Original and Predicted Prices:")
display(df)
```

### 4. Visualizing the Regression Line

Visualizing the data points and the fitted regression line helps us understand how well the model captures the relationship between house size and price.

```python
# Visualize the data and the regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Size (m²)', y='Price ($1000s)', data=df, s=100, label='Actual Prices')
plt.plot(df['Size (m²)'], df['Predicted Price ($1000s)'], color='red', linestyle='--', label='Regression Line')
plt.title('House Price Prediction with Linear Regression')
plt.xlabel('Size (m²)')
plt.ylabel('Price ($1000s)')
plt.grid(True)
plt.legend()
plt.show()
```
