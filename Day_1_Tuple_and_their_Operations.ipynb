# Day 1: Python Tuples for AI, NLP, LLMs & Agentic AI

## What is a Tuple?

A **tuple** is an ordered, immutable (cannot be changed after creation) collection of items in Python. Tuples are used to store **fixed data** that should remain constant throughout a program.

**Syntax:**

my_tuple = (item1, item2, item3)


**Key Features**

* Ordered collection
* Immutable (cannot add, remove, or modify elements)
* Allows duplicate values
* Faster than lists for fixed data
* Can store different data types

---

# 1. Creating a Tuple

```python
numbers = (10, 20, 30, 40, 50)

print(numbers)
```

### AI Application

Store fixed mathematical constants or model parameters that should never change.

---

# 2. Tuple of AI Models

```python
models = (
    "Linear Regression",
    "Decision Tree",
    "Random Forest",
    "Transformer"
)

print(models)
```

### Application

Store a fixed list of supported AI models.

---

# 3. Accessing Tuple Elements

```python
frameworks = (
    "NumPy",
    "Pandas",
    "PyTorch",
    "TensorFlow"
)

print(frameworks[0])
print(frameworks[2])
print(frameworks[-1])
```

### Application

Access predefined frameworks used in an AI project.

---

# 4. Tuple Packing

```python
student = ("Ali", 24, "AI")

print(student)
```

### Application

Store fixed student information or experiment metadata.

---

# 5. Tuple Unpacking

```python
name, age, field = ("Ali", 24, "AI")

print(name)
print(age)
print(field)
```

### Application

Retrieve experiment information efficiently.

---

# 6. Loop Through a Tuple

```python
languages = (
    "Python",
    "R",
    "Julia",
    "Java"
)

for language in languages:
    print(language)
```

### Application

Display supported programming languages.

---

# 7. Tuple Length

```python
datasets = (
    "IMDB",
    "SQuAD",
    "Common Crawl",
    "WikiText"
)

print(len(datasets))
```

### Application

Count the number of available datasets.

# Why Use Tuples Instead of Lists in AI?

| List                     | Tuple                        |
| :----------------------- | :--------------------------- |
| Mutable                  | Immutable                    |
| Can add/remove items     | Cannot add/remove items      |
| Slightly slower          | Slightly faster              |
| Used for dynamic data    | Used for fixed data          |
| Example: Dataset samples | Example: Model configuration |

---
