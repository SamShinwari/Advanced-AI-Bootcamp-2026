# Day 1: Python Lists for AI, NLP, LLMs & Agentic AI

A **list** is one of the most important Python data structures. In AI and machine learning, lists are used to store datasets, sentences, tokens, predictions, embeddings, batches, and training results.

## 1. Creating a List

````python
numbers = [10, 20, 30, 40, 50]

print(numbers)
````

**Output**

```
[10, 20, 30, 40, 50]
```

**AI Application:**
Store feature values or numerical data.

## 2. List of Sentences (NLP)

````python
sentences = [
    "I love AI",
    "Python is amazing",
    "Large Language Models are powerful"
]

print(sentences)
````

**Application:**
Each sentence can become a training example for an NLP model.

## 3. Accessing Elements

````python
sentences = [
    "Hello",
    "Welcome",
    "Python"
]

print(sentences[0])
print(sentences[1])
print(sentences[-1])
````

**Application:**
Access individual text samples from a dataset.

## 4. Adding New Data

````python
dataset = []

dataset.append("Artificial Intelligence")
dataset.append("Machine Learning")
dataset.append("Deep Learning")

print(dataset)
````

**Application:**
Continuously collect new training examples.

## 5. Removing Data

````python
languages = ["Python", "Java", "C++", "Rust"]

languages.remove("Java")

print(languages)
````

**Application:**
Remove corrupted or unwanted samples.

## 6. Loop Through Dataset

````python
dataset = [
    "AI",
    "ML",
    "NLP",
    "LLM"
]

for item in dataset:
    print(item)
````

**Application:**
Read every training sample before preprocessing.

# Real AI Applications

## Example 1: Tokenization

````python
sentence = "I love Artificial Intelligence"

tokens = sentence.split()

print(tokens)
````

**Output**

```
['I', 'love', 'Artificial', 'Intelligence']
```

Lists store the tokens before they are converted into IDs.

## Example 2: Vocabulary

````python
vocabulary = [
    "AI",
    "Machine",
    "Learning",
    "Python"
]

print(vocabulary)
````

Used for creating a vocabulary before assigning token IDs.

## Example 3: Mini Batch Training

````python
batch = [
    "Sentence 1",
    "Sentence 2",
    "Sentence 3",
    "Sentence 4"
]

for sentence in batch:
    print(sentence)
````

Deep learning models process one **batch (list)** at a time.

## Example 4: Storing Predictions

````python
predictions = []

predictions.append("Positive")
predictions.append("Negative")
predictions.append("Positive")

print(predictions)
````

Store model outputs for evaluation.

## Example 5: Chatbot Conversation History (LLMs)

````python
conversation = []

conversation.append("User: Hello")
conversation.append("Assistant: Hi! How can I help?")
conversation.append("User: Explain Transformers")

for message in conversation:
    print(message)
````

LLMs maintain conversation history as a list of messages.

## Example 6: Prompt List

````python
prompts = [
    "Summarize this article.",
    "Translate to Pashto.",
    "Generate Python code."
]

for prompt in prompts:
    print(prompt)
````

Useful for batch prompting and prompt engineering.

## Example 7: RAG Documents

````python
documents = [
    "Document about AI",
    "Document about NLP",
    "Document about Transformers"
]

print(documents)
````

A Retrieval-Augmented Generation (RAG) system stores retrieved documents in a list before sending them to the LLM.

## Example 8: Multi-Agent AI

````python
agents = [
    "Research Agent",
    "Coding Agent",
    "Reviewer Agent"
]

for agent in agents:
    print(agent)
````

Agentic AI frameworks maintain a list of collaborating agents.

## Example 9: Training Loss

````python
losses = [0.95, 0.74, 0.52, 0.31, 0.15]

print("Final Loss:", losses[-1])
````

Store the loss after each training epoch.

## Example 10: Embeddings

````python
embedding = [0.12, -0.43, 0.89, 0.56, -0.11]

print(embedding)
````

Although real embeddings usually use NumPy or PyTorch tensors, they can first be understood as a list of numbers.
