## Part 15: Building a Simple Neural Network

### Input

We define the input features for a single sample:
- Age = 20
- Experience = 2

This is represented as a column vector `X`:

```python
import numpy as np

X = np.array([
    [20],
    [2]
])

print("Input vector X:")
print(X)
```

### Weights

These are the weights `W` connecting the input layer to the hidden layer (or output layer in this simple case). Each row corresponds to a neuron, and each column corresponds to an input feature.

```python
W = np.array([
    [0.4, 0.8],
    [0.7, 0.2]
])

print("Weight matrix W:")
print(W)
```

### Bias

The bias `b` is added to the weighted sum of inputs for each neuron.

```python
b = np.array([
    [1],
    [1]
])

print("Bias vector b:")
print(b)
```

### Neural Network Computation

The neural network computes the output `Y` using the formula:
`Y = WX + b`

```python
Y = W @ X + b

print("Computed output Y:")
print(Y)
```

### Step-by-Step Calculation

Let's break down the calculation for each neuron:

**First neuron:**
- `(0.4 × 20) + (0.8 × 2) = 8 + 1.6 = 9.6`
- `9.6 + bias (1) = 10.6`

**Second neuron:**
- `(0.7 × 20) + (0.2 × 2) = 14 + 0.4 = 14.4`
- `14.4 + bias (1) = 15.4`

### Final Output

Based on the calculations, the final output `Y` is:

```python
print("Final Output Y:")
print(Y)
```

## Part 16: ReLU Activation Function

Neural networks often apply an activation function after the weighted sum. One of the most common is the Rectified Linear Unit (ReLU).

### Formula:

`ReLU(x) = max(0, x)`

This function keeps positive values unchanged and replaces negative values with zero.

### Example Input:

Consider the following input vector `Y`:

```python
import numpy as np

Y_relu_input = np.array([
    [-3],
    [5],
    [-1],
    [8]
])

print("Input for ReLU function:")
print(Y_relu_input)
```

### ReLU Computation:

We apply the ReLU function to the input vector. This involves taking the maximum of 0 and each element in the input.

```python
relu_output = np.maximum(0, Y_relu_input)

print("Output after applying ReLU:")
print(relu_output)
```

### Final Output:

The final output after applying the ReLU activation function is:

```python
print("Final ReLU Output:")
print(relu_output)
```
