```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

AND  gate dataset

```python
# Input features
X = np.array([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
])

# Target labels
y = np.array([
    [0.],
    [0.],
    [0.],
    [1.]
])
```

##build perceptron model

```python
model = Sequential([
    Dense(
        units=1,
        activation='sigmoid',
        input_shape=(2,)
    )
])
```

###compile the model

```python
model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

###train the model

```python
history = model.fit(
    X,
    y,
    epochs=1000,
    verbose=0
)
```

###model evaulation

```python
loss, accuracy = model.evaluate(X, y, verbose=0)

print("Loss:", loss)
print("Accuracy:", accuracy)
```

###model prediction

```python
predictions = model.predict(X)

print("Probabilities:")
print(predictions)

print("\nBinary Predictions:")
print((predictions >= 0.5).astype(int))
```

### view learned weight

```python
weights, bias = model.layers[0].get_weights()

print("Weights:")
print(weights)

print("\nBias:")
print(bias)
```
