Single layer perceptron model using iris dataset
using scikit library

import libraries

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

load the iris dataset

```python
def load_and_prepare_data():
    """Loads the Iris dataset and prepares features and binary target."""
    iris = load_iris()
    X = iris.data
    # Binary target: Setosa = 0, Versicolor + Virginica = 1
    y = (iris.target != 0).astype(int)
    return X, y
```

split the data into train and test split

```python
def split_data(X, y, test_size=0.2, random_state=42):
    """Splits the data into training and testing sets."""
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        random_state=random_state
    )
    return X_train, X_test, y_train, y_test
```

use perceptron model

```python
def train_perceptron_model(X_train, y_train, random_state=42):
    """Creates and trains a Perceptron model."""
    model = Perceptron(random_state=random_state)
    model.fit(X_train, y_train)
    return model
```

makes predictions
evaulates model accuracy

```python
def evaluate_model(model, X_test, y_test):
    """Makes predictions and evaluates the model's accuracy."""
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)
    return accuracy
```

model flow

```python
if __name__ == '__main__':
    X, y = load_and_prepare_data()
    X_train, X_test, y_train, y_test = split_data(X, y)
    model = train_perceptron_model(X_train, y_train)
    evaluate_model(model, X_test, y_test)
```
