### Python Dictionaries: Real-World Applications

A dictionary is a versatile data structure in Python, ideal for storing collections of data that need to be accessed by a unique identifier (a key). In real-world applications, dictionaries are commonly used for tasks such as:

*   **User Profiles**: Storing user-specific information (name, email, settings) using their user ID or username as the key.
*   **Product Catalogs/Inventory**: Managing product details (name, price, stock) with product IDs or SKUs as keys.
*   **Configuration Settings**: Storing application or game settings (e.g., `difficulty: 'hard'`, `volume: 75`).
*   **API Responses**: Parsing and handling data received from web services, which often come in JSON format (which maps directly to Python dictionaries).

Let's explore these concepts with practical examples.

### 1. Creating a Dictionary: User Profiles & Product Inventory

In real applications, you often create dictionaries to represent structured data. For instance, an empty dictionary might represent a new, uninitialized record, while filled dictionaries can hold user profiles or product details.

```python
new_user_profile = {}
print(f"1.1 New User Profile (empty): {new_user_profile}")

product_inventory = {
    101: {'name': 'Laptop', 'price': 1200.00, 'stock': 50},
    102: {'name': 'Mouse', 'price': 25.00, 'stock': 150},
    103: {'name': 'Keyboard', 'price': 75.00, 'stock': 80}
}
print(f"1.2 Product Inventory (by ID): {product_inventory}")

user_settings = {'admin': {'theme': 'dark', 'notifications': True, 'language': 'en_US'}}
print(f"1.3 User Settings (by username): {user_settings}")

customer_order = {'order_id': 'AX123', 'items': ['Laptop', 'Mouse'], 'total_amount': 1225.00, 'is_paid': False}
```

The `dict()` constructor is useful when you have data in other formats, like a list of key-value pairs or when defining dictionaries with simple string keys as keyword arguments.

```python
app_config = dict(database='PostgreSQL', environment='production', debug_mode=False)
print(f"1.5 Application Configuration: {app_config}")
```

### 2. Accessing Elements: Retrieving User Data & Product Details

In a real-world application, you'll frequently need to retrieve specific pieces of information. For example, getting a user's email from their profile or checking the price of a product.

```python
user_profile = {'username': 'johndoe', 'email': 'john.doe@example.com', 'is_active': True}
product_details = {101: {'name': 'Laptop', 'price': 1200.00, 'stock': 50}}

print("\n--- 2. Accessing Elements ---")
print(f"2.1 User's Email: {user_profile['email']}")
print(f"2.2 Product Name (ID 101): {product_details[101]['name']}")
```

```python
user_profile = {'username': 'johndoe', 'email': 'john.doe@example.com', 'is_active': True}

print(f"2.4 User's Email (using .get()): {user_profile.get('email')}")
print(f"2.5 User's Phone (using .get(), default None): {user_profile.get('phone')}")
print(f"2.6 User's Phone (using .get(), default 'Not Provided'): {user_profile.get('phone', 'Not Provided')}")
```

### 3. Modifying Elements: Updating User Data, Stock Levels, & Cart Management

Real-world applications constantly modify data. This could involve updating a user's subscription status, changing a product's price, or adding/removing items from a shopping cart.

```python
user_data = {'id': 'user_001', 'name': 'Jane Doe', 'subscription_status': 'basic'}

print("\n--- 3. Modifying Elements ---")
user_data['preferred_language'] = 'French'
print(f"3.1 User data after adding 'preferred_language': {user_data}")
```

```python
user_data = {'id': 'user_001', 'name': 'Jane Doe', 'subscription_status': 'basic', 'preferred_language': 'French'}

user_data['subscription_status'] = 'premium'
print(f"3.2 User data after updating 'subscription_status': {user_data}")
```

```python
user_data = {'id': 'user_001', 'name': 'Jane Doe', 'subscription_status': 'premium', 'preferred_language': 'French'}

del user_data['preferred_language']
print(f"3.3 User data after deleting 'preferred_language': {user_data}")
```

### 4. Built-in Dictionary Functions/Methods: Data Management & Utility

These methods are crucial for performing common data management tasks, from getting a list of all product IDs to merging configuration files or preparing temporary data structures.

```python
shopping_cart = {'item_A': 2, 'item_B': 1, 'item_C': 3}
print(f"Current shopping cart: {shopping_cart}")

print(f"4.1 Number of unique items in cart: {len(shopping_cart)}")
```

```python
product_database = {'P001': 'Laptop', 'P002': 'Monitor', 'P003': 'Webcam', 'P004': 'Microphone'}
product_ids = product_database.keys()
print(f"4.2 Product IDs: {list(product_ids)}")
```

```python
product_database = {'P001': 'Laptop', 'P002': 'Monitor', 'P003': 'Webcam', 'P004': 'Microphone'}
product_names = product_database.values()
print(f"4.3 Product Names: {list(product_names)}")
```

```python
user_permissions = {'admin': 'full_access', 'editor': 'partial_access', 'viewer': 'read_only'}
permissions_list = user_permissions.items()
print(f"4.4 User Permissions (role, access): {list(permissions_list)}")
```

```python
current_order = {'sku_1': 2, 'sku_2': 1, 'sku_3': 4}
removed_quantity = current_order.pop('sku_2')
print(f"4.5 Quantity of 'sku_2' removed: {removed_quantity}, Order now: {current_order}")
```

```python
last_login_sessions = {'user_A': '2023-10-26 10:00', 'user_B': '2023-10-26 11:30', 'user_C': '2023-10-26 12:45'}
last_session_removed = last_login_sessions.popitem()
print(f"4.6 Last login session removed: {last_session_removed}, Remaining sessions: {last_login_sessions}")
```

```python
default_settings = {'theme': 'light', 'notifications': True}
user_preferences = {'notifications': False, 'font_size': 'medium'}

default_settings.update(user_preferences)
print(f"4.7 Combined user settings: {default_settings}")

default_settings.update(debug_mode=False, log_level='INFO')
print(f"4.7 Combined settings after additional updates: {default_settings}")
```

```python
temporary_cache = {'data_id_1': 'cached_value_A', 'data_id_2': 'cached_value_B'}
temporary_cache.clear()
print(f"4.8 Temporary cache after clear(): {temporary_cache}")
```

```python
system_config = {'server_ip': '192.168.1.1', 'port': 8080, 'timeout': 30}
modified_config = system_config.copy()
print(f"4.9 Original config: {system_config}, Copied config: {modified_config}")
modified_config['port'] = 9000
print(f"4.9 Original config (unchanged): {system_config}, Modified copy: {modified_config}")
```

```python
new_user_ids = ['user_101', 'user_102', 'user_103']
user_statuses = dict.fromkeys(new_user_ids, 'pending_activation')
print(f"4.10 Initial user statuses: {user_statuses}")
```

```python
report_filters = {'status': 'active', 'category': 'electronics'}

print(f"4.11 Current 'status' filter: {report_filters.setdefault('status', 'all')}")
print(f"4.11 Current 'region' filter: {report_filters.setdefault('region', 'APAC')}")
print(f"4.11 Report filters after setdefault: {report_filters}")
```

### 5. Iterating Through a Dictionary: Processing Data & Generating Reports

Iterating through dictionaries is essential for displaying data, processing records, or generating reports in applications. You might want to list all users, show products in a category, or summarize order items.

```python
user_scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78}

print("\n--- 5. Iterating Through a Dictionary ---")

print("5.1 Listing user names (default iteration over keys):")
for user in user_scores:
    print(f"  User: {user}")

print("5.2 Listing user names explicitly (.keys()):")
for user_key in user_scores.keys():
    print(f"  User ID/Name: {user_key}")
```

```python
user_scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78}

print("5.3 Listing user scores (.values()):")
for score in user_scores.values():
    print(f"  Score: {score}")
```

```python
product_prices = {'P001': 1200, 'P002': 25, 'P003': 75}

print("5.4 Displaying product IDs and prices (.items()):")
for product_id, price in product_prices.items():
    print(f"  Product ID: {product_id}, Price: ${price:.2f}")

print("\n--- End of Real-World Dictionary Examples ---")
```
